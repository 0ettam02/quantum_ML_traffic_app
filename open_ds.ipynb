# Visualizzazione un biflusso del dataset completo 

```python
import pickle
import sys
from pprint import pprint
import numpy as np

path = "mirage2019_LOPEZ_lopez_lopez_36P_4F_APP_xST_PAD_metadata.pickle"

example_idx = 2000
n_pkts = 36
n_features = 4 # DIR, PL, TCPWIN, IAT

with open(path, "rb") as f:
    biflows = pickle.load(f)
    labels = pickle.load(f)

# come è fatto un biflusso
pprint(biflows[example_idx][:n_pkts,:n_features])

# a che app e attività appartiene quel biflusso
pprint(labels[example_idx])

# App e attività del dataset
print(f"Unique apps: {np.unique(labels)}")
```
