# Verifica somma predizione è uguale al valore reale 100k

```python
import pandas as pd

df = pd.read_csv('output/predictions.dat', delim_whitespace=True, header=None)

df['sum'] = (df[0] == df[1]).astype(int)
print("la somma di tutte le predizioni corrette è:", df['sum'].sum(), "su un totale di", len(df), "predizioni")

totale = len(df) - df['sum'].sum()
print("la somma di tutte le predizioni errate è:", totale)

accuracy = df['sum'].sum() / len(df)
print("l'accuracy del modello è:", accuracy)

df.head()
```
