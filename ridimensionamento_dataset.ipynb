# Shaping dataset originale

```python
import pickle
import numpy as np

path = "mirage2019_LOPEZ_lopez_lopez_36P_4F_APP_xST_PAD_metadata.pickle"

with open(path, "rb") as f:
    biflows = pickle.load(f)
    labels = pickle.load(f)

print(f"Dataset originale: {len(biflows)} campioni, primo campione shape: {biflows[0].shape}")
print(f"Numero di etichette: {len(labels)}")
```

# Ridimensionamento dataset originale in dataset con solo 10K valori

```python
import random

n_samples = 20000
random.seed(42)  

indices = random.sample(range(len(biflows)), n_samples)

biflows_sampled = [biflows[i] for i in indices]
labels_sampled = [labels[i] for i in indices]

print(f"Dataset campionato: {len(biflows_sampled)} campioni")


path_sampled = "mirage2019_LOPEZ_20k_sampled.pickle"

with open(path_sampled, "wb") as f:
    pickle.dump(biflows_sampled, f)
    pickle.dump(labels_sampled, f)

print("Dataset campionato salvato correttamente!")
```
